# Graph Anomaly Detection In Tree Structured Networks

In this notebook we will specifically look at the task of anomaly detection within tree structured networks. The idea behind this exploration is to see what type of structural and syntactic information the model can hold. Namely, trees have a very defined set of "rules" that define their type (e.g. balanced, connected, etc.). By training on regular trees of different types and then testing on trees with one or more properties broken, the idea is to explore what the model can begin to detect.

```python
# Fun some setup for the project
# Silence annoying pytorch deprecated warnings
import warnings
warnings.filterwarnings("ignore")

from train import *
from create_graphs import *
%matplotlib inline

print ('CUDA Available:', torch.cuda.is_available())
```

```python
# Generate the model datasets (i.e. the dataset used to train
# the model - normal data)
args_tree, train_tree, val_tree, test_tree = get_graph_data("tree", isModelDataset=True)
# Save the max_previous node to allow for model 
# compatability on future datasets
max_prev_node = args_tree.max_prev_node
```

```python
# Initialize the dataset and dataloader for analyzing the nlls. 
# Note that we use batch_size = 1 because we want the nll for each
# data point not an average across a batch.
# We should also consider using Graph_sequence_Sampler_pytorch_nll.
# This sampler expands the size of the dataset by for each graph 
# creating many different bfs permutations. The idea behind this sampler
# is to test the models permutation invariance. 

train_dataset = Graph_sequence_sampler_pytorch_rand(train_tree,max_prev_node=args_tree.max_prev_node,max_num_node=args_tree.max_num_node)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=1, num_workers=args_tree.num_workers)

test_dataset = Graph_sequence_sampler_pytorch_rand(test_tree, max_prev_node=args_tree.max_prev_node,max_num_node=args_tree.max_num_node)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=1, num_workers=args_tree.num_workers)
```

```python
# Model initialization
# Using GraphRNN

rnn = GRU_plain(input_size=args_tree.max_prev_node, embedding_size=args_tree.embedding_size_rnn,
                        hidden_size=args_tree.hidden_size_rnn, num_layers=args_tree.num_layers, has_input=True,
                        has_output=True, output_size=args_tree.hidden_size_rnn_output).to(device)
output = GRU_plain(input_size=1, embedding_size=args_tree.embedding_size_rnn_output,
                        hidden_size=args_tree.hidden_size_rnn_output, num_layers=args_tree.num_layers, has_input=True,
                        has_output=True, output_size=1).to(device)
```

```python
# Let's do some analysis on the training data
nlls, avg_nlls = calc_nll(args_tree, train_loader, rnn, output, max_iter=50,load_epoch=500)
```

```python
# Plot un averaged nlls (over the feature dimensions of the graph sequence) for the model's training dataset
plt.figure()
sns.distplot(nlls,kde=True)
```

The above result makes some sense because of the dramatically different sizes of the trees. For example, we have a 2-ary tree of height 3 and a 4-ary tree of height 4. Therefore, if we are just summing the loglikelihoods across the feature dimension there will be a large magnitude imbalance between the small trees and large trees. Even if the small tree have several unlikely features, it is much more likely that the sum of a few large numbers will be less than the sum of many medium numbers.

```python
# Plot averaged nlls (over the feature dimensions of the graph sequence) for the model's training dataset
plt.figure()
sns.distplot(avg_nlls,kde=True)
```

```python
# Analysis of the test data set nlls.
test_nlls, test_avg_nlls = calc_nll(args_tree, test_loader, rnn, output, max_iter=50, load_epoch=500)
```

```python
# Un-averaged nlls over the test normal distribution
plt.figure()
sns.distplot(test_nlls, kde=True)
plt.title('NLL summed across features')
```

```python
# Averaged nlls over test normal distribution
plt.figure()
sns.distplot(test_avg_nlls, kde=True)
plt.title('NLL averaged across features')
```

# Discussion/thoughts
These results do not seem super interesting at first. However, I believe that they highlight a real issue pertaining to the number of nodes in different graphs. We see that in the un-averaged graph it does highlight some potential issues. Specifically, we see that the scale of the loglikelihood is heavily dependant on the number of nodes in the graph, which makes sense if we are summing across the features (Note, this represents the true loglikilihood, where we originally have the sum over individual 'chain-rule' components in the sequence). The anomalous dataset contains graphs with fewer and more nodes than the training dataset, and we see that especially for those with many more nodes the relative loglikelihood is massive, demonstrating the insignificance of the scale of the summed likelihood. Looking at the second graph of the averaged likelihoods we uncover another issue, in that when we average across a long sequence (i.e. many nodes) a lot of fine grained information is lost. Specifically, for the trees that have heights greater than those in the training dataset (5, 6), these trees have valid subtrees that existed exatly in the training dataset. For example, the excluding the last layer in a height five tree is exactly the height for tree that exists in the training dataset. Therefore, it seems likely that many individual "tokens/decisions" in the graph rnn sequence may have high likelihoods, while a few key additions (i.e. nodes that make the tree too tall) make the entire graph unlikely. However, when we just average across all the graphical structure it becomes much harder to tease out these unlikely events. Therefore, it seems like an important point of exploration is to look at the distriubion of likelihoods across the "features/tokens" in a single graph sequence.

## 2) Removing edges

For this exploration we will consider what happens if we remove one or more edges at random from the tree. Because our training dataset only included full r-ary trees this should cause issues, especially in the case that a portion of the tree becomes disconnected.  

```python
args_edge, graphs_edge = get_graph_data("tree_r_edge_1", isModelDataset=False)

# Note that instead of passing in args.max_prev_node, we pass in the saved max_prev_node
# - hack to allow for GraphRNN to work on any dataset
edge_dataset = Graph_sequence_sampler_pytorch_rand(graphs_edge,max_prev_node=max_prev_node,max_num_node=args_edge.max_num_node)
edge_loader = torch.utils.data.DataLoader(edge_dataset, batch_size=1, num_workers=args_edge.num_workers)
```

```python
# Note we pass in the args_tree because the model we are
# using are trained in the enzyme dataset
edge_nlls, edge_avg_nlls = calc_nll(args_edge, edge_loader, rnn, output, max_iter=50, 
                                        load_epoch=500, train_dataset="tree")
```

```python
# un-averaged nlls over the ladder normal distribution
plt.figure()
sns.distplot(edge_nlls, kde=True)
plt.title('NLL un-averaged across features')
```

```python
# Averaged nlls over the ladder normal distribution
plt.figure()
sns.distplot(edge_avg_nlls, kde=True)
plt.title('NLL averaged across features')
```
